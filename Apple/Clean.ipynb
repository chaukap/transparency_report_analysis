```python
import pandas as pd
```

```python
df = pd.read_csv("Raw_Data/account_requests.csv")
df.head()
```

```python
df = df[df["Country/Region"] == "United States of America"]
df["half"] = [1 if "H1" in period else 2 for period in df["TR Period"]]
df["year"] = [period[0:4] for period in df["TR Period"]]

df.rename(columns={
    "TR Period": "period",
    "Country/Region": "country",
    "Account Requests Received": "total_requests",
    "Accounts Specified in Requests": "total_requests_accounts",
    "Requests Challenged in Part or Rejected in Full": "total_requests_challenged",
    "Requests where Only Non-Content Data Provided": "only_non_data_provided",
}, inplace=True)

reduced_names = [
    "period",
    "country",
    "total_requests",
    "total_requests_accounts",
    "total_requests_challenged",
    "only_non_data_provided",
    "year",
    "half"
]

df = df[reduced_names]
df["country"] = "United States"

df.head()
```

```python
df.to_csv("../Clean_Data/Apple.csv", index=False)
```
